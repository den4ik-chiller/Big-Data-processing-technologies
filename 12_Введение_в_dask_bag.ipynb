```python
with open('Dostoevskiy Fedor. Igrok - BooksCafe.Net.txt', 'r') as file:
    text = file.read()
    sentences = nltk.sent_tokenize(text)
for sent in sentences:
    if len(sent) < 10:
        print(sent)
```

4. На основе файла `addres_book.json` создайте `dask.bag`. Посчитайте количество мобильных и рабочих телефонов в наборе данных

```python
import dask.bag as db
import json
with open('addres-book.json', 'r') as f:
    data = json.load(f)
addres_bag = db.from_sequence(data)
mobile = addres_bag.pluck('phones').flatten().pluck('phone').filter(lambda x: '+' in x).count().compute()
work = addres_bag.pluck('phones').flatten().pluck('phone').filter(lambda x: '+' not in x).count().compute()
print(mobile, work)
```

2. Модифицируйте функцию разбора JSON таким образом, чтобы в каждый словарь c информацией об отзыве добавить ключ `rating`. Значение получите на основе названия файла (см. аргумент `include_path`), использовав для этого регулярное выражение.

```python
import json
import glob
import re

def parse_review_json_with_rating(json_string, include_path):
    json_obj = json.loads(json_string)
    match = re.search(r'reviews_(\d+)\.json', include_path)
    if match:
        rating = int(match.group(1))
        json_obj['rating'] = rating
    return json_obj
```

```python
# b = db.read_text('reviews_full/*.json').map(parse_review_json_with_rating, include_path='reviews_3.json')
# b.take(3)
```

```python
# Читаем все файлы в список
files = glob.glob('reviews_full/*.json')

# Создаем первый dask.bag
b_ratings = db.read_text(files[0]).map(parse_review_json_with_rating, include_path=files[0])

# Объединяем остальные dask.bag
for file in files[1:]:
    cur = db.read_text(file).map(parse_review_json_with_rating, include_path=file)
    b_ratings = db.concat([b_ratings, cur])

b_ratings.take(3)
```

4. Отфильтруйте `bag`, сохранив только отзывы, оставленные в 2014 и 2015 годах.

```python
# импортируем модуль datetime для работы с датами
from datetime import datetime

# функция, которая проверяет, был ли отзыв оставлен в 2014 или 2015 году
def is_review_in_2014_or_2015(review):
    year = datetime.strptime(review['date'], '%Y-%m-%d').year
    return year in [2014, 2015]

# отфильтровываем отзывы, оставленные в 2014 и 2015 годах
reviews_2014_2015 = b_ratings.filter(is_review_in_2014_or_2015)
reviews_2014_2015.take(3)
```

5. Выполните препроцессинг отзывов:
    * привести строки к нижнему регистру
    * обрезать пробельные символы в начале и конце строки
    * удалите все символы, кроме английских букв и пробелов
    
Примените препроцессинг ко всем записям из `bag`, полученного в задании 4.

```python
def preprocess_review(review_dict):
    # нижний регистр и обрезка пробельных символов
    review_dict['review'] = review_dict['review'].lower().strip()
    # удаляем все символы, кроме английских букв и пробелов
    review_dict['review'] = re.sub('[^a-zA-Z\s]', '', review_dict['review'])
    return review_dict

preprocessed_reviews = reviews_2014_2015.map(preprocess_review)
preprocessed_reviews.take(3)
```

7. Посчитайте, как часто в наборе, полученном в задании 5, встречается та или иная оценка

```python
ratings = preprocessed_reviews.pluck('rating')
freq = ratings.frequencies().compute()
print(freq)
```

8. Найдите среднее значение `rating` в выборке

```python
total_ratings = sum([rating * count for (rating, count) in freq])
total_counts = sum([count[1] for count in freq])
average_rating = total_ratings / total_counts
average_rating
```

9. Используя метод `foldby`, подсчитать максимальную длину отзывов в зависимости от оценки `rating` в наборе, полученном в задании 5.

```python
max_review_lengths = preprocessed_reviews.foldby('rating', lambda acc, x: max(acc, len(x['review'])), initial=0).compute()
```
